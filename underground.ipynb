```python
import csv
import random
from random import randint
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
from itertools import combinations
import random
import numpy as np
```

```python
file = csv.reader("метро1.csv", delimiter = ",")
matrix_1 = []
matrix_2 = []
with open("метро1.csv") as file:
    reader = csv.reader(file, delimiter=',', quotechar=',')
    for row in reader:
        matrix_1.append(row)
g = nx.Graph()
t=[matrix_1[i][0] for i in range(len(matrix_1))]
```

```python
s = [29, 26, 23, 13, 12, 24, 23, 8, 12, 25, 23, 6, 14, 7, 4]
n=-2
k = 0
for i in range(len(s)):
    n = n + s[i] 
    for j in range(k,n):
        g.add_edge(t[j],t[j+1],weigth=random.randint(3,7))
    k = s[i] + 1
g.add_edge(t[len(t)-2],t[len(t)-1],weigth=random.randint(3,7))
```

```python
f = plt.figure(figsize=(30,30))
print(g.nodes())
len(g.nodes)
pos = nx.spring_layout(g)
nx.draw_networkx(g, pos)
plt.title("Random Graph Generation Example")
plt.show()
```
